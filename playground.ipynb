```python
import pandas as pd
```

```python
# Чтение .dat-файлов

HH_data = pd.read_csv('conditions/HH.dat', header=None, sep='  ')
HH_data.sample(2)
```

```python
# Чтение .grd-файлов

with open('conditions/HXY1.grd', 'r') as file:
    data = file.readlines()
HXY1_data = [str[:-2].split(sep='  ') for str in data[5:]]
HXY1_data = pd.DataFrame(HXY1_data)
HXY1_data.sample(2)
```

```python
# HH_data.to_excel('HH.xlsx')
# HXY1_data.to_excel('HXY1.xlsx')
HH_data.to_csv('HH.csv')
HXY1_data.to_csv('HXY1.csv')
```
